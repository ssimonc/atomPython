### ZIT

Gode and Sunder in 1993 made it clear that a certain number of behaviours observable in a real market are not due to the rationality of its participants but appear even with participants behaving randomly. They thus introduced the ZIT, Zero Intelligence Trader, which is a random agent: with only an interval of possible prices, an interval of quantity to trade and possibly a probability of trading, these agents randomly choose trades to make.

```python
class ZITTrader_v2(Trader):
    def __init__(self, market, initial_assets=None, cash=0):
        Trader.__init__(self, market, initial_assets, cash)
    def __str__(self):
        return "ZITv2 " + super().__str__()
    def decide_order(self, market, asset):
        # We're counting the number of orders placed in the orderbooks by this agent 
        nb_placed_orders = sum(len([o for o in market.orderbooks[asset2].asks.tree+market.orderbooks[asset2].bids.tree if o.source == self]) for asset2 in market.orderbooks.keys())
        if nb_placed_orders == 0:
            direction = random.choice(['ASK', 'BID'])
            if direction == 'ASK' and self.assets[asset] > 0: # If we have quantity > 0 of shares for this asset... 
                # You cannot sell a quantity of stock > to the one you have.
                return LimitOrder(asset, self, 'ASK', random.randint(1000, 9999), random.randint(1, min(self.assets[asset], 9)))
            elif direction == 'BID' and self.cash >= 1000: # To buy, you need money... 
                price = random.randint(1000, min(self.cash, 9999))
                # We don't buy for volume of assets with a total price > to the cash we have.
                return LimitOrder(asset, self, 'BID', price, random.randint(1, min(self.cash//price, 9)))
        return None # In all other cases, no order is placed.
    
class ZITPeriodique(Trader):
    def __init__(self, market, initial_assets=None, cash=0, periode=1):
        Trader.__init__(self, market, initial_assets, cash)
        self.periode = periode
    def __str__(self):
        return "ZITdelta %i" % self.trader_id
    def decide_order(self, market, asset):
        
        if ((market.time+1)%self.periode==0):
            return LimitOrder(asset, self, random.choice(['ASK', 'BID']), random.randint(1000,9999), random.randint(1, 9))
        else:
            return None 
```

These agents are originally based on a client-trader model: the client asks the trader to place orders at a price v and the trader tries to maximize his profit in relation to that price and the market. Atom does not follow this model and the price v asked by the client is therefore simulated by estimating the fundamental value of the product.

### ZIC : Zero-Intelligence Constrained Traders

These traders only buy at higher random prices and only sell at random prices below the value of v.

```python
# v = hash table of redemption value i.e. dic of perceived fundamental values
class ZIC(Trader):
    def __init__(self, market, initial_assets=None, cash=0, v=5000):
        Trader.__init__(self, market, initial_assets, cash)
        self.v = v
    def __str__(self):
        return "ZIC " + super().__str__()
    def decide_order(self, market, asset):
        nb_placed_orders = len([o for o in market.orderbooks[asset].asks.tree+market.orderbooks[asset].bids.tree if o.source == self])
        if nb_placed_orders == 0:
            direction = random.choice(['ASK', 'BID'])
            if direction == 'ASK':
                return LimitOrder(asset, self, 'ASK', random.randint(self.v[asset], 9999), random.randint(1, 9))
            elif direction == 'BID':
                return LimitOrder(asset, self, 'BID', random.randint(1000, self.v[asset]), random.randint(1, 9))
        return None
```

### ZIP : Zero-Intelligence-Plus Traders

These traders optimize their profit margin according to the last price thanks to learning: if it is close to v, the margin will be reduced (because an order proposed at a price too far away from v will not find a taker), otherwise it will be increased (because an order proposed at a price too close to v prevents potential profits). 

```python
# v = redemption value i.e. perceived fundamental value
class ZIPab(Trader):
    def __init__(self, market, initial_assets=None, cash=0, v=5000, momentum=0.5, beta=0.3):
        Trader.__init__(self, market, initial_assets, cash)
        self.v = v
        self.price = -1
        self.marginA = 0
        self.prev_changeA = 0
        self.marginB = 0
        self.prev_changeB = 0
        self.momentum = momentum # momentum [0 - 1] i.e. coeff of importance of past change
        self.beta = beta #  learning rate [0.1 - 0.5]
    def __str__(self):
        return "ZIPab " + super().__str__()
    def decide_order(self, market, asset):
        
        def update_margin(direction):
            P = market.prices_hist[asset]
            if len(P)>=1:
                q = P[-1]
                if self.price == -1:
                    self.price = q
                diff = q - self.price
                
                if (direction=='ASK'):
                    change = ((1.0-self.momentum)*(self.beta*diff)) + (self.momentum*self.prev_changeA) 
                    self.prev_changeA = change
                    new_marg = ((self.price + change)/self.v[asset]) - 1.0
                    self.marginA = new_marg
                else:
                    change = ((1.0-self.momentum)*(self.beta*diff)) + (self.momentum*self.prev_changeB)
                    self.prev_changeB = change
                    new_marg = ((self.price + change)/self.v[asset]) - 1.0
                    self.marginB = new_marg
        
        
        nb_placed_orders = len([o for o in market.orderbooks[asset].asks.tree+market.orderbooks[asset].bids.tree if o.source == self])
        if nb_placed_orders == 0:
            direction = random.choice(['ASK', 'BID'])
            update_margin(direction)
            if direction=='ASK':
                #print('ASK - margin:',self.marginA)
                return LimitOrder(asset, self, direction, int(self.v[asset]*(1+self.marginA)), random.randint(1, 9))
            else:
                #print('BID - margin:',self.marginB)
                return LimitOrder(asset, self, direction, int(self.v[asset]*(1+self.marginB)), random.randint(1, 9)) 
        return None
    
    
class ZIP(Trader):
    def __init__(self, market, initial_assets=None, cash=0, v=5000, momentum=0.5, beta=0.3):
        Trader.__init__(self, market, initial_assets, cash)
        self.v = v
        self.price = -1
        self.margin = 0
        self.prev_change = 0
        self.margin
        self.prev_change
        self.momentum = momentum # momentum [0 - 1] i.e. coeff of importance of past change
        self.beta = beta #  learning rate [0.1 - 0.5]
    def __str__(self):
        return "ZIP " + super().__str__()
    def decide_order(self, market, asset):
        
        def update_margin():
            P = market.prices_hist[asset]
            if len(P)>=1:
                q = P[-1]
                if self.price == -1:
                    self.price = q
                diff = q - self.price
                change = ((1.0-self.momentum)*(self.beta*diff)) + (self.momentum*self.prev_change) 
                self.prev_change = change
                new_marg = ((self.price + change)/self.v[asset]) - 1.0
                self.margin = new_marg
        
        
        nb_placed_orders = len([o for o in market.orderbooks[asset].asks.tree+market.orderbooks[asset].bids.tree if o.source == self])
        if nb_placed_orders == 0:
            direction = random.choice(['ASK', 'BID'])
            update_margin()
            return LimitOrder(asset, self, direction, int(self.v[asset]*(1+self.margin)), random.randint(1, 9))
        return None
```

### GDX
  
In 2002, Tesauro and Bredin created an agent based on a belief function, i.e. the function that gives the probability that an order will be fulfilled. For a given direction, the agent looks at the history of exchanges and deduces its belief function. If most of the sales before time t have been made very close to a price p', it is useless to look for a large profit with a price p of sale much higher than p' or or a price of purchase much lower than p': the probability that the order will be filled is very low. It is therefore preferable to send an order at a price p close to p', even if it means sacrificing profit. To calculate the exact value of p, it is easy from the belief function and v to deduce the expectation of each possible price. The price with the maximum expectation corresponds to the price that is both close enough to the last prices so that the chance that the order will find a taker is great and far enough away to minimize the compromise with respect to profit. This price calculation is relaunched at each turn of the agent's speech according to a history that is often partial (n last prices fixed).

```python
P = m.prices_hist["Apple"]

def f_belief(market,asset,p,direction):
    P = market.prices_hist[asset]
    aag = 0
    abl = 0
    for price in P:
        if price>=p:
            aag+=1
        else:
            abl+=1
    book = market.orderbooks[asset]
    asks = book.asks.tree
    bids = book.bids.tree
    ual = 0
    for ask in asks:
        if ask.price <= p:
            ual+=1
    ubg = 0
    for bid in bids:
        if bid.price >= p:
            ubg+=1
    if direction=='ASK':
        if (2*aag + ubg + ual)==0:
            return 0
        return ((2*aag + ubg) / (2*aag + ubg + ual))
    else:
        if (2*abl + ual + ubg)==0:
            return 0
        return ((2*abl + ual) / (2*abl + ual + ubg))
```

```python
class GDX(Trader):
    def __init__(self, market, initial_assets=None, cash=0, v=5000):
        Trader.__init__(self, market, initial_assets, cash)
        self.v = v
    def __str__(self):
        return "GDX " + super().__str__()
    def decide_order(self, market, asset):
        directionAbs = {'ASK':1,'BID':-1}
        
        
        nb_placed_orders = len([o for o in market.orderbooks[asset].asks.tree+market.orderbooks[asset].bids.tree if o.source == self])
        if nb_placed_orders == 0:
            direction = random.choice(['ASK', 'BID'])
            
            expected_surplus = lambda p : directionAbs[direction]*(p-self.v[asset])*f_belief(market,asset,p,direction) # probability price p of a certain direction is accepted 
            #print("res",expected_surplus(p),"f",f_belief(market,asset,p,direction),"p",p,"l",self.v[asset],direction)
            
            prices = np.linspace(3000,7000,400) # for 400 prices between 3000 and 7000 
            exp_s  = np.array([expected_surplus(p) for p in prices]) # we calculate the profit expectation using the belief function
            p = prices[np.argmax(exp_s)] # We get the prize that has the maximum expectation 
            return LimitOrder(asset, self, direction, p, random.randint(1, 9)) # limitorder at this price
        return None
```

### AA : Adaptative Aggressive traders

Vytelingum, Cliff and Jennings proposed the Adaptive Aggressive (AA) agent. This agent relies on automatic learning to adjust a parameter of aggressiveness. This parameter varies over time and adapts according to the orders accepted in the history. The agent strategy is therefore somewhere between a very aggressive strategy where the agent will take risks until buying or selling at a loss (depending on its fixed price) or be passive and try to trade with the highest profit, regardless of the probability that the order will find takers. <img src="agg.png" alt="drawing" width="500"/>  The curves in this figure show the possible prices according to different aggressiveness values.   which represents the learning frequency of the aggressiveness. A low value corresponds to an accentuation of aggressiveness and is adapted to a fast market, for example one that has just been launched or that is experiencing a strong and sudden price variation. A high value, on the other hand, corresponds to a decrease in aggression and is suitable for a stable market.

```python
# weighted moving average 
# for N last prices
# T is the last price
# each weight is worth rho times what the one after that is worth
# 0<rho<1 so each weight is worth less as it gets further in the price history from the last price
# weights are normalised so sum is 1

def get_weights(N_last_prices,rho=0.9):
    w = np.empty(N_last_prices)
    w[-1] = 1
    for i in range(2,N_last_prices+1):
        w[N_last_prices-i] = w[N_last_prices-i+1]*rho
    w = w/sum(w)
    return w

def get_wma(market,asset,N_last_prices,rho):
    P = market.prices_hist[asset]
    if len(P)>N_last_prices:
        H = P[-N_last_prices:]
    else:
        H = P
        N_last_prices = len(H)
    w = get_weights(N_last_prices,rho)
    wma = sum(w*H)
    return wma

print(get_wma(m,'Apple',10,0.9))
print(get_wma(m,'Apple',100,1))
print(get_wma(m,'Apple',100,.9))


```
